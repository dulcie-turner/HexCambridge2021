#Introduction

Heat transfer between a solid surface and a fluid has many engineering applications, including, heat exchangers, boilers and condensers. We call natural convection the heat transfer enhanced by gravitational effects (the change in density of a gas, such as a space radiator), and forced convection when a mechanical device (such as a fan) induces the flow. In both cases, heat transfer rates are usually calculated using a heat transfer coefficient, $h$, which is defined such that,
\begin{equation}
Q = hA\Delta T \qquad (1)
\end{equation}
where $Q$ is the heat transfer rate, $A$ is the surface area,
and $\Delta T$ is the temperature difference between the surface and  surrounding fluid. The units of $h$ are $\rm Wm^{-2}K^{-1}$.

This experiment examines forced convection, in which air flow is
induced by a fan, and natural convection, in which air flow is induced by  temperature differences, for a small diameter (25$\rm \mu m$) electrically  heated wire. Radiation is neglected throughout, but you should bear in mind that radiative heat transfer increases with the fourth power of absolute
temperature.

#Experimental setup


Figure 2 shows the experimental arrangement. A platinum wire is located near the entry of a 20 mm diameter perspex tube which has a smooth intake section. Air is drawn through the tube by means of a fan, and controlled by means of a valve. Flow rates are measured using a variable section flow meter, which connects the vertical position of a float to the flow rate.
The platinum wire forms one limb of the Wheatstone bridge circuit shown in Fig. 1(b). The circuit is supplied from a 12 V lead-acid battery, and the current may be varied using the variable resistor, $R_v$. Electrical heating of the wire results in an increase in its temperature and hence its resistance, $R_w$. Equilibrium is reached when the electrical energy supplied
balances heat loss to the air in accordance with equation (1).

The resistances in the upper branch of the Wheatstone bridge ($R_1$ and $R_2$) are approximately 10,000 times greater than those in the lower branch ($R_3$ and $R_w$), so that nearly all the current registered by the ammeter passes through the platinum wire. By adjusting the current it is possible to balance the bridge such that the bridge voltmeter, $V_1$, reads zero. The ratios of the resistors in the upper and lower branches are then the same ( i.e., $R_w/R_3 = R_2/R_1$). $R_w$ varies according to: 
\begin{equation}R_w/R_o = 1 + \kappa \Delta T \qquad (2)\end{equation}      

where $\kappa$ is a constant, so that a constant $R_w$ means a constant wire temperature, and thus $\Delta T$. 



#Experimental parameters



The physical parameters and ambient conditions used in the experiments are described as follows.

```python
d = 25.0E-6       # wire diameter (m)
kappa = 3.5E-3    # K^-1
ell = 0.020       # wire length (m)
Vbatt = 12        # battery voltage (V)
R0 = 5            # wire resistance (Ohm)
pa = (765/760)*101324.7    # atmospheric pressure (Pa)
Ta = 20.5         # air temperature (C)
RJa = 287          # gas constant for air (J/kg K)
g = 9.81          # gravitational constant (m/s^2)

def lambdaa(T) :   # air conductivity (W/ m K ) for air temperatures in kelvin          
  lambdaa = 0.02624*(T/300)**0.8646
  return lambdaa 

def mua(T) :   # air dynamic viscosity (Pa-s) for air temperatures in kelvin          
  mua = 1.458E-6*T**1.5/(T + 110.4) 
  return mua

def rhoa(T) :   # air density (kg/m^3) for air temperatures in kelvin          
  rhoa = pa/(RJa*T) 
  return rhoa

def Ua(FF) : # air velocity (m/s) as a function of float reading F (cm)
   Ua = 0.09457*FF + 1.314
   return Ua

# source: https://onlinelibrary.wiley.com/doi/pdf/10.1002/9780470516430.app2
```

<H1> Forced flow </H1>

Under steady, conditions, the heat dissipated through the wire is given by: 

$Q = I_2^2 R_w \qquad (3)$

but under the current conditions where $R_w+R_3 << R_1 + R2$, $I_2 = I$, which is measured. The resistance is calculated from the measured voltage as: 

$R_w = V_2/I_2 \approx V_2/I \qquad (4)$

The total heat dissipated must be equal to the heat generated, so that: 

$Q = h A \Delta T = I^2 R_w \qquad (5)$

We are interested in the non-dimensional heat transfer rate per unit area of the wire. We consider it in comparison to the estimated heat transfer rate if conduction were the only mechanism of heat transfer, which is approximated as: 

$Q_c = \lambda A \frac{\Delta T}{d} \qquad (6)$

where $\lambda$ is the thermal conductivity of air. So the non-dimensional heat transfer rate, called the Nusselt number, is: 

$Nu = \frac{Q}{Q_c} = \frac{h d}{\lambda} \qquad (7)$

We obtain $h$ and therefore $Nu$ from the measured total heat transfer, the wire resistance, the dimensions of the wire (length $\ell$ and diameter $d$), and properties of air: 

$Nu = \frac{I^2 R_w d}{\lambda A \Delta T} = \frac{I^2 R_w}{\lambda \pi \ell \Delta T} \qquad (8)$. 

We can estimate the properties of air either at ambient conditions, or more accurately, at the intermediate temperature of the film between the wire surface and ambient given by: 

$T_f = T_a + \Delta T/2 \qquad (9)$

where the temperature rise $\Delta T$ is obtained from the relationship between the resistance and temperature rise, Eq. (2).






> Relationship between Nu and Re: 

As we observe from everyday phenomena (wind chill for example), increasing the velocity also increases the heat transfer. This happens because the increased convection sharpens the temperature gradients near the surface, thus increasing the heat transfer. 

Let us assume that the important input variables to the heat transfer problem are geometry ($d$), operating conditions (velocity $U$), and fluid properties (density $rho$, viscosity $\mu$, specific heat $c_p$, conductivity $\lambda$), and that the output is the heat transfer coefficient per unit area ($h$). An exercise in dimensional analysis using the Buckingham-$\pi$ theorem connecting input and output variables; 
\begin{equation}
h  = f(\rho, \mu, c_p, \lambda, d, U)
\end{equation}
yields a reduced set of non-dimensional numbers after dimensions are collected: 7 variables, and 4 dimensions (mass, length, time, temperature) lead to 3 non-dimensional numbers, so that: 
\begin{equation}
Nu = \frac{h d}{\lambda}  = g(\frac{\rho U d}{\mu},\frac{\mu c_p}{\lambda}) = g (Re, Pr)
\end{equation}
where $Re = \frac{\rho U d}{\mu}$ and $Pr =\frac{\mu c_p}{\lambda}$ Other combinations are possible, but these are the popular ones.

# Natural Convection 

For natural convection, we have the same equations as before, but now, the flow is induced by the heating of the wire itself, which lowers the density of the gas, resulting in its upward motion. 

We can calculate the same Nusselt number, but now the independent variable is the temperature (or density) of the gas leading to the flow. The flow is now dependent on the rate of expansion of the gas as the temperature increases, characterized by the thermal expansivity $\beta = \frac{1}{v}\left(\frac{\partial v}{\partial T}\right)_p$, where $v$ is specific volume, and which is equal to $1/T$ for a perfect gas), the gravitational constant $g$, along with the previous variables listed for heat and momentum transfer, i.e.  $d$, $\rho$, $\mu$, $\lambda$, and the specific heat $c_p$.

Following a similar exercise in non-dimensional analysis as for forced convection, but now instead of the velocity $U$, we have gravity $g$, and the new expansion parameter $\beta$, the analysis yields: 
\begin{equation}
Nu = \frac{h d}{\lambda}  = \phi(\frac{\rho d^3\beta \Delta T g}{\mu\lambda/(\rho c_p)}) 
\end{equation}
where the Rayleigh number is 
\begin{equation}
  Ra = \frac{\rho d^3\beta \Delta T g}{\mu\lambda/(\rho c_p)} = Gr Pr
\end{equation}
where $Gr = \frac{d^3 g\Delta T/T }{\nu^2}$. In the case of air, $Pr \approx 0.7$ for a wide range of conditions. 

> Note: For the correlations of natural convection, the properties should be evaluated at *ambient conditions*, and $\Delta T/T = \Delta T/T_a$ .

# Start you calculations and answers from here onwards

```python
import math
from scipy.stats import linregress
Ta = Ta + 273
lowRe = []
log_lowRe = []
lowNu = []
log_lowNu = []
lowTemp = []
for i in range(len(FFL)):
  m = mua(Ta)
  density = rhoa(Ta)
  U = Ua(FFL[i]/10)
  Re = (density * U * d) / m
  logRe = math.log10(Re)
  lowRe.append(Re)
  log_lowRe.append(logRe)
  Rw = V2L[i]/IL[i]
  lamda = lambdaa(Ta)
  changeT = ((Rw/R0)-1)/kappa
  Nu = ((IL[i])**2 * Rw) / (lamda * math.pi *changeT * ell)
  lowNu.append(Nu)
  log_lowNu.append(math.log(math.log10(Nu)))
  lowTemp.append(changeT)
plt.plot(log_lowNu, log_lowRe)
plt.xlabel("Re number")
plt.ylabel("Nu number")
plt.show()
slope, intercept, r_value, p_value, std_err = linregress(log_lowNu, log_lowRe)
A = 10**intercept
a = slope
print("A = ", A, "a = ", a)

HighRe = []
log_HighRe = []
HighNu = []
log_HighNu = []
HighTemp = []
for i in range(len(FFH)):
  m = mua(Ta)
  density = rhoa(Ta)
  U = Ua(FFH[i]/10)
  Re = (density * U * d) / m
  HighRe.append(Re)
  log_HighRe.append(math.log10(Re))
  Rw = V2H[i]/IH[i]
  lamda = lambdaa(Ta)
  changeT = ((Rw/R0)-1)/kappa
  Nu = ((IH[i])**2 * Rw) / (lamda * math.pi *changeT * ell)
  HighNu.append(Nu)
  log_HighNu.append(math.log(math.log10(Nu)))
  HighTemp.append(changeT)
plt.plot(log_HighNu, log_HighRe)
plt.xlabel("Re number")
plt.ylabel("Nu number")
plt.show()
slope, intercept, r_value, p_value, std_err = linregress(log_HighNu, log_HighRe)
A = 10**intercept
a = slope
print("A = ", A, "a = ", a)

nNu = []
nRa = []
nTemp = []
pr = 0.7
for i in range(len(V2N)):
  Rw = V2N[i]/IN[i]
  changeT = ((Rw/R0)-1)/kappa
  v = mua(Ta)/rhoa(Ta)
  gr = ((d**3) * g * changeT) / (Ta * (v**2))
  Ra = gr * pr
  nRa.append(Ra)
  lamda = lambdaa(Ta)
  Nu = ((IN[i])**2 * Rw) / (lamda * math.pi *changeT * ell)
  nNu.append(Nu)
  nTemp.append(changeT)
plt.plot(nNu, nRa)
plt.xlabel("natural Ra number")
plt.ylabel("natural Nu number")
plt.show()

plt.plot(lowNu, lowTemp )
plt.xlabel("delta Temperature")
plt.ylabel("Nu number")
plt.title("forced convection for low R2")
plt.show()

plt.plot(HighNu, HighTemp )
plt.xlabel("delta temperature")
plt.ylabel("Nu number")
plt.title("forced convection for High R2")
plt.show()

plt.plot(nNu, nTemp )
plt.xlabel("delta temperature")
plt.ylabel("Nu number")
plt.title("natural convection")
plt.show()

#forced convection if better for values og high h
```
